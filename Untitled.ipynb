```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import yfinance as yf

# 1. Télécharger les données
df = yf.download("AAPL", start="2010-01-01", end="2024-01-01")
data = df[['Close']].dropna()

# 2. Normaliser les données
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

# 3. Fonction pour créer les séquences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

SEQ_LEN = 60
X, y = create_sequences(data_scaled, SEQ_LEN)

# 4. Split train/test
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 5. Créer le modèle LSTM amélioré
model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(SEQ_LEN, 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# 6. Entraînement
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1, verbose=1)

# 7. Prédictions
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

# 8. Inverser la normalisation
y_train_inv = scaler.inverse_transform(y_train.reshape(-1, 1))
train_pred_inv = scaler.inverse_transform(pred_train)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
test_pred_inv = scaler.inverse_transform(pred_test)

# 9. MSE
mse_train = mean_squared_error(y_train_inv, train_pred_inv)
mse_test = mean_squared_error(y_test_inv, test_pred_inv)

# 10. Directional Accuracy
direction_pred = np.sign(test_pred_inv[1:] - test_pred_inv[:-1])
direction_true = np.sign(y_test_inv[1:] - y_test_inv[:-1])
directional_accuracy = np.mean(direction_pred == direction_true)

# 11. Plot complet
plt.figure(figsize=(16, 6))
plt.plot(np.arange(len(y_train_inv)), y_train_inv, label='Training Actual', color='blue')
plt.plot(np.arange(len(train_pred_inv)), train_pred_inv, label='Training Forecast', color='cyan', linestyle='--')
plt.plot(np.arange(len(y_train_inv), len(y_train_inv) + len(y_test_inv)), y_test_inv, label='Testing Actual', color='green')
plt.plot(np.arange(len(train_pred_inv), len(train_pred_inv) + len(test_pred_inv)), test_pred_inv, label='Testing Forecast', color='red', linestyle='--')
plt.title("LSTM Stock Forecasting - AAPL")
plt.xlabel("Time Steps")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 12. Zoom sur le dernier mois
plt.figure(figsize=(12, 5))
plt.plot(y_test_inv[-30:], label='Actual (Last 30 days)', color='green')
plt.plot(test_pred_inv[-30:], label='Predicted (Last 30 days)', color='red', linestyle='--')
plt.title("Zoom Forecast - Last 30 Days")
plt.xlabel("Days")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 13. Affichage des métriques
print(f"Training MSE: {mse_train:.4f}")
print(f"Testing MSE: {mse_test:.4f}")
print(f"Directional Accuracy (testing): {directional_accuracy * 100:.2f}%")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import yfinance as yf

# Download Apple data
df = yf.download("AAPL", start="2010-01-01", end="2024-01-01")
data = df[['Close', 'Volume']].dropna()

# Normalize data
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

# Sequence creation
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length, 0])  # target is 'Close'
    return np.array(X), np.array(y)

SEQ_LEN = 60
X, y = create_sequences(data_scaled, SEQ_LEN)

# Train-test split
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Model
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(SEQ_LEN, X.shape[2])),
    Dropout(0.2),
    LSTM(32, return_sequences=False),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_reduce = ReduceLROnPlateau(monitor='val_loss', patience=3, factor=0.5, verbose=1)

# Fit
history = model.fit(X_train, y_train, validation_split=0.1,
                    epochs=30, batch_size=32, verbose=1,
                    callbacks=[early_stop, lr_reduce])

# Predictions
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

# Inverse transform predictions
y_train_inv = scaler.inverse_transform(np.hstack((y_train.reshape(-1, 1),
                                                   X_train[:, -1, 1].reshape(-1, 1))))[:, 0]
train_pred_inv = scaler.inverse_transform(np.hstack((pred_train,
                                                      X_train[:, -1, 1].reshape(-1, 1))))[:, 0]
y_test_inv = scaler.inverse_transform(np.hstack((y_test.reshape(-1, 1),
                                                  X_test[:, -1, 1].reshape(-1, 1))))[:, 0]
test_pred_inv = scaler.inverse_transform(np.hstack((pred_test,
                                                     X_test[:, -1, 1].reshape(-1, 1))))[:, 0]

# Metrics
mse_train = mean_squared_error(y_train_inv, train_pred_inv)
mse_test = mean_squared_error(y_test_inv, test_pred_inv)

def directional_accuracy(y_true, y_pred):
    return np.mean((np.diff(y_true) > 0) == (np.diff(y_pred) > 0)) * 100

da_test = directional_accuracy(y_test_inv, test_pred_inv)

# Plot Full
plt.figure(figsize=(16, 8))
plt.plot(y_test_inv, label='Testing Actual', color='green')
plt.plot(test_pred_inv, label='Testing Forecast', color='red', linestyle='--')
plt.title("AAPL Stock Price Prediction (Full View)")
plt.xlabel("Time Steps")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot Zoomed (Last Month)
plt.figure(figsize=(14, 6))
plt.plot(y_test_inv[-30:], label='Actual (Last 30)', color='green')
plt.plot(test_pred_inv[-30:], label='Predicted (Last 30)', color='red', linestyle='--')
plt.title("AAPL Stock Price Prediction - Zoomed (Last Month)")
plt.xlabel("Time Steps")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print Metrics
print(f"Training MSE: {mse_train:.4f}")
print(f"Testing MSE: {mse_test:.4f}")
print(f"Directional Accuracy (testing): {da_test:.2f}%")
```
